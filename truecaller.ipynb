```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time

driver = webdriver.Chrome(ChromeDriverManager().install())

driver.get('https://www.truecaller.com/')
driver.maximize_window()

sign_in_button = driver.find_element_by_link_text('Sign in')
sign_in_button.click()

time.sleep(2)

google_button = driver.find_element_by_link_text('MICROSOFT')
google_button.click()

time.sleep(2)

email_input = driver.find_element_by_name('loginfmt')
email = ""              #enter microsoft id
email_input.send_keys(email)
email_input.click()

time.sleep(2)

submit_button_id = driver.find_element_by_id("idSIButton9")
submit_button_id.click()

time.sleep(2)

password_input = driver.find_element_by_name("passwd")
password = ""           #enter micrsoft password
password_input.send_keys(password)
password_input.click()

time.sleep(2)

submit_button_pass = driver.find_element_by_id("idSIButton9")
submit_button_pass.click()

time.sleep(2)

stay_sign = driver.find_element_by_id("idBtn_Back")
stay_sign.click()

time.sleep(2)

search_input = driver.find_element_by_css_selector('input[data-v-541c74f4=""]')
phone_number = ""        #enter mobile number
search_input.send_keys(phone_number)
search_input.click()

time.sleep(2)

search_button = driver.find_element_by_css_selector("button[data-v-541c74f4]")
search_button.click()

time.sleep(10)

name = driver.find_element_by_css_selector("div[data-v-6b8b5293]").text
email = driver.find_element_by_css_selector("div[data-v-2579f823]").text
print("Registerd: Yes")
print(name)
print(email)

driver.quit()
```
